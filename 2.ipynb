```python
import boto3
import json
import s3fs

# Function to load the secrets from the local file
def load_secrets(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)
secrets = load_secrets('secrets.json')
```

```python
aws_access_key_id = secrets['AWS_ACCESS_KEY_ID']
aws_secret_access_key = secrets['AWS_SECRET_ACCESS_KEY']
aws_region = secrets['AWS_REGION']
```

```python
import boto3
import io
import pandas as pd
import json

aws_id = secrets['AWS_ACCESS_KEY_ID']
aws_secret = secrets['AWS_SECRET_ACCESS_KEY']
bucket_name = 'greencoop-airbyte'
object_key = "greencoop-airbyte/Station_amateur_Weather_Underground_a_Ichtegem,_Belgique/2025_03_14_1741979874105_0.jsonl"


s3 = boto3.client('s3', aws_access_key_id=aws_id, aws_secret_access_key=aws_secret)
obj = s3.get_object(Bucket=bucket_name, Key=object_key)
data = obj['Body'].read()
df = pd.read_json(io.BytesIO(data), lines=True)
```

```python
import magic

mime = magic.Magic(mime=True)
file_type = mime.from_buffer(data)
print("Detected file type:", file_type)
```
